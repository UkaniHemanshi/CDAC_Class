    <b>3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
</b>
</div>  

```python
data = open('./DataSet/test.txt','r')
lines= data.readlines()
xlst = []
ylst = []
for line in lines:
    x = int(line[0])
    y = int(line[2])
    xlst.append(x)
    ylst.append(y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line plot with text file")
plt.xlim(left=min(xlst), right=max(xlst)) 
plt.ylim(bottom=min(ylst), top=max(ylst)) 
plt.plot(xlst, ylst, c='r') 
plt.show()
```

axis values.</b>
</div>

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])
plt.plot(x, y, marker='o')

# Display Current Limits
current_xlim = plt.gca().get_xlim()      # GCA = Get Current Axis Limits
current_ylim = plt.gca().get_ylim() 
print(f'Current x-axis limits: {current_xlim}')
print(f'Current y-axis limits: {current_ylim}')

# Display updated limits
plt.xlim(0, 6)  
plt.ylim(0, 12)  

new_xlim = plt.gca().get_xlim()
new_ylim = plt.gca().get_ylim()
print(f'New x-axis limits: {new_xlim}')
print(f'New y-axis limits: {new_ylim}')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Display and Set Axis Limits')

plt.grid()
plt.show()
```

<div class = "alert alert-block alert-info"
    <b>15. Write a Python program to create multiple plots.</b>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Sample data
x = np.linspace(0, 10, 100)                 
y4 = np.exp(-x)               

# Create a figure with gridspec
fig = plt.figure(figsize=(12, 10))
gs = GridSpec(2, 3, height_ratios=[1, 0.5])  # 50% for the first, 50% for the rest

# Plot the first graph (occupying 50% of the figure)
ax1 = fig.add_subplot(gs[0, :])  # Span across all columns in the first row
ax1.set_title("Multi-Line plot with Csv_dataframe")

plt.xlabel("Date")
plt.ylabel("Values")

ax1.plot(df['Date'], df['Open'], c='b')
ax1.plot(df['Date'], df['High'], c='orange')
ax1.plot(df['Date'], df['Low'], c='g')
ax1.plot(df['Date'], df['Close'], c='r')
ax1.legend(df[['Open','High','Low','Close']])
ax1.grid()


# Plot the second graph (occupying 17% - left column)
ax2 = fig.add_subplot(gs[1, 0])
ax2.set_title("Simple curve and plot some quantities")

plt.xlabel("Date")
plt.ylabel("Values")

ax2.plot(df['Date'], df['Close'], 'b--',marker='o')
plt.xticks(rotation=45)


# Plot the third graph (occupying 17% - middle column)
ax3 = fig.add_subplot(gs[1, 1])

ax3.set_title("Scatter Plot of Quantities with X and Y Positions")
x1 = np.linspace(2,10,5)
x2 = np.linspace(2,8,5)

y = np.array([22,25,23,20,21])
z = np.array([40,25,10,20,30])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

ax3.scatter(x1,y,linewidth=2,c='orange')
ax3.scatter(x2,z,linewidth=1,c='r')

# Plot the fourth graph (occupying 17% - right column)
ax4 = fig.add_subplot(gs[1, 2])
ax4.plot(x, y4, color='b', label='Exponential Decay')
ax4.set_title('Exponential Decay')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
ax4.legend()
ax4.grid()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

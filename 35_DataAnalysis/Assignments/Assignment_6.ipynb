```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./DataSet/alphabet_stock_data.csv')
df
```

<div class = "alert alert-block alert-info"
    <b>1. Write a Pandas program to create a line plot of the historical stock prices of 
Alphabet Inc. between two specific dates.</b>
</div>   

```python
start_date = '2020-04-01'
end_date = '2020-05-03'

plt.figure(figsize=(18,8))

filtered_date = df[(df['Date']>=start_date) & (df['Date']<=end_date)]
plt.plot(filtered_date['Date'],filtered_date['Close'],ls='--',c='b',marker='o',label='Closing Price')
plt.xticks(rotation= 90)
plt.xlabel('Dates',fontsize=15,c='red',labelpad=15)
plt.ylabel('Closing Price',fontsize=15,c='red',labelpad=15)
plt.title('Line plot of the historical stock prices btn two specific dates',fontsize=25,c='green',pad=10)
plt.grid(True)
plt.legend(fontsize=15,loc='lower right')
plt.show()
```

<div class = "alert alert-block alert-info"

prices of Alphabet Inc. between two specific dates.</b>

```python
start_date = '2020-04-01'
end_date = '2020-05-03'

plt.figure(figsize=(18,8))

filtered_date = df[(df['Date']>=start_date) & (df['Date']<=end_date)]
plt.plot(filtered_date['Date'],filtered_date['Close'],ls='--',c='r',marker='o',label='Closing Price')
plt.plot(filtered_date['Date'],filtered_date['Open'],ls=':',c='g',marker='.',label='Opening Price')
plt.xticks(rotation= 90)
plt.xlabel('Dates',fontsize=15,c='red',labelpad=15)
plt.ylabel('Price',fontsize=15,c='red',labelpad=15)
plt.title('Line plot of the Closing and Opening price of historical stock price btn two specific dates',fontsize=25,c='green',pad=10)
plt.grid(True)
plt.legend(fontsize=15,loc='lower right')
plt.show()
```

```python
start_date = '2020-04-01'
end_date = '2020-05-03'

plt.figure(figsize=(18,8))

filtered_date = df[(df['Date']>=start_date) & (df['Date']<=end_date)]
plt.bar(filtered_date['Date'],filtered_date['Volume'],ls='--',color='pink',label='Volume')

plt.xticks(rotation= 90)
plt.xlabel('Dates',fontsize=15,c='red',labelpad=15)
plt.ylabel('Volume',fontsize=15,c='red',labelpad=15)
plt.title('Bar plot of the trading volume',fontsize=25,c='green',pad=10)
plt.grid(True)
plt.legend(fontsize=15,loc='upper left')
plt.show()
```

<div class = "alert alert-block alert-info"
    <b>4. Write a Pandas program to create a bar plot of opening, closing stock prices of 
Alphabet Inc. between two specific dates.</b>
</div> 

```python
start_date = '2020-04-01'
end_date = '2020-05-03'

plt.figure(figsize=(18,8))

filtered_date = df[(df['Date']>=start_date) & (df['Date']<=end_date)]
x = np.arange(len(filtered_date['Date']))

bar_width = 0.4
plt.bar(x - bar_width/2, filtered_date['Open'], width = bar_width, color='#FF7F50', label='Opening Price')
plt.bar(x + bar_width/2, filtered_date['Close'], width = bar_width, color='#00BFFF', label='Closing Price')

plt.xticks(rotation= 90)
plt.xlabel('Dates',fontsize=15,c='red',labelpad=15)
plt.ylabel('Price',fontsize=15,c='red',labelpad=15)
plt.title('Bar plot of opening, closing stock prices of Alphabet' ,fontsize=25,pad=10)

plt.xticks(x + bar_width / 2, filtered_date['Date'])
plt.legend(fontsize=15,loc='upper left')
plt.show()
```

<div class = "alert alert-block alert-info"
    <b>7. Write a Pandas program to create a histograms plot of opening, closing, high, 
low stock prices of Alphabet Inc. between two specific dates.</b>
</div> 

```python
plt.figure(figsize=(12, 8))

start_date = '2020-04-01'
end_date = '2020-04-20'

filtered_date = df[(df['Date']>=start_date) & (df['Date']<=end_date)]
x = np.arange(len(filtered_date['Date']))

# Opening prices
plt.subplot(2, 2, 1)
plt.hist(filtered_date['Open'], bins=10, color='blue', alpha=0.7)

# Closing prices
plt.subplot(2, 2, 1)
plt.hist(filtered_date['Close'], bins=10, color='green', alpha=0.5)

# High prices
plt.subplot(2, 2, 1)
plt.hist(filtered_date['High'], bins=10, color='orange', alpha=0.4)

# Low prices
plt.subplot(2, 2, 1)
plt.hist(filtered_date['Low'], bins=10, color='red', alpha=0.2)
plt.yticks(x + bar_width / 2, filtered_date['Date'])
plt.title('Histogram plot for Open, Close, High, Low ')
plt.xlabel('Price')
plt.ylabel('Date')

# Show the plots
plt.tight_layout()
plt.show()
```

<div class = "alert alert-block alert-info"
    <b>8. Write a Pandas program to create a stacked histograms plot of opening, 
closing, high, low stock prices of Alphabet Inc. between two specific dates.</b>
</div> 

```python
# Prepare data for stacking
prices = pd.DataFrame({
    'Open': filtered_date['Open'],
    'Close': filtered_date['Close'],
    'High': filtered_date['High'],
    'Low': filtered_date['Low']
})
x = np.arange(len(filtered_date['Date']))
# Plotting stacked histogram
plt.figure(figsize=(12, 8))

# Use hist with the stacked parameter
prices.plot.hist(stacked=True, bins=10, alpha=0.7)

# Adding titles and labels
plt.title('Stacked Histogram of Stock Prices for Alphabet Inc.')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.yticks(x + bar_width / 2, filtered_date['Date'])
plt.legend(title='Price')

# Show the plot
plt.tight_layout()
plt.show()    
```

<div class = "alert alert-block alert-info"
    <b>9. Write a Pandas program to draw a horizontal and cumulative histograms plot 
of opening stock prices of Alphabet Inc. between two specific dates.
</b>
</div> 

```python
# Prepare data for stacking
prices = pd.DataFrame({
    'Open': filtered_date['Open'],
})

# Plotting stacked histogram
plt.figure(figsize=(12, 8))

# Use hist with the stacked parameter
prices.plot.hist(orientation = "horizontal",cumulative=True)

# Adding titles and labels
plt.title('Cumulative Histogram of Stock Prices for Alphabet Inc.')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

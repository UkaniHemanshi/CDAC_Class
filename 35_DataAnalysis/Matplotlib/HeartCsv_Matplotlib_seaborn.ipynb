```python
# line plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
ages = df.groupby('age').median().reset_index()
df
```

```python
df.head()
```

```python
df.info()
```

## Using Pandas

```python
ages.plot.line('age','chol')
plt.show()
```

## Using Matplotlib

```python
fig,ax = plt.subplots()
ax.plot(ages["age"],ages["chol"],ls = ':',lw = 1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
plt.show()
```

## Using Seaborn

```python
sns.lineplot(x ="age",y = "chol",data=ages,ls=":",lw=1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
# Another index example - Air BNB dataset from kaggle
import pandas as pd
```

```python
df_air = pd.read_csv('AB_NYC_2019.csv')
df_air.head(4)
```

```python
df2 = df_air.set_index('id')
df2.head(3)
```

```python
df2.info()
```

```python
numeric_columns = df2.select_dtypes(include = ['float64','int64'])
numeric_columns['room_type'] = df2['room_type']   # add room_type column from df2 to numeric_columns
numeric_columns.head(4)
```

```python
df3 = numeric_columns.groupby('room_type').mean()
df3
```

```python
pd.set_option('display.max_column',None)  # give all the column and None means it set no restriction bcz pandas sometime restrict some column by default
df3
```

```python
df3 = df3.reset_index()
pd.reset_option('display.max_columns')
df3
```

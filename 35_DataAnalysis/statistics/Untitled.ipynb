```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SP500 oil gold bitcoin.csv')
df.head()
```

```python
corr_bit_gold = df[['BITCOIN','Gold']].corr()
corr_bit_sp500 = df[['BITCOIN','S&P500']].corr()
```

```python
covar_bit_gold = df[['BITCOIN','Gold']].cov()
covar_bit_sp500 = df[['BITCOIN','S&P500']].cov()
```

```python
print('correlation btn bitcoin and gold')
print(corr_bit_gold)
```

```python
print('correlation btn bitcoin and sp500')
print(corr_bit_sp500)
```

```python
print('covariance btn bitcoin and gold')
print(covar_bit_gold)
```

```python
print('covariance btn bitcoin and sp500')
print(covar_bit_sp500)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].scatter(df['BITCOIN'],df['Gold'],color='blue',alpha=0.5)
axes[0].set_title(f"bit_gold correlation:{corr_bit_gold.loc['BITCOIN','Gold']:.2f}")
axes[0].set_xlabel('BITCOIN')
axes[0].set_ylabel('Gold')
axes[0].grid(True)


axes[1].scatter(df['BITCOIN'],df['S&P500'],color='red',alpha=0.5)
axes[1].set_title(f"bit_sp500 correlation:{corr_bit_sp500.loc['BITCOIN','S&P500']:.2f}")
axes[1].set_xlabel('BITCOIN')
axes[1].set_ylabel('S&P500')
axes[1].grid(True)

xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()


test_x = xmin + 0.7 * (xmax - xmin)
test_y = ymin + 0.3 * (ymax - ymin)

plt.text(test_x,test_y,f"correlation: {corr_bit_gold.loc['BITCOIN','Gold']:.2f}",ha= "center",va = 'center',fontsize = 12)
plt.grid(True)
plt.show()


plt.tight_layout()
plt.show()
```
